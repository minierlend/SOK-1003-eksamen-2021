```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def kredit():
    laanemengde = input('Hvor mye vil du låne? \n')
    laanemengde = float(laanemengde.replace(' ', ''))

    maaneder = input('Hvor mange måneder vil du ha lånet? \n')
    maaneder = int(maaneder)


    rente = input('Hva er renten på lånet ditt? \n')
    rente = float(rente.replace(',', '.').replace('%', ''))

    periodegebyr = input('Hvor mye er periodegebyret på? ? \n')
    periodegebyr = float(periodegebyr)

    laan = laanemengde * ( 1 + rente/100 ) + periodegebyr * maaneder
    
    x_akse = np.linspace(0, maaneder, int(laan))
    y_akse = laanemengde * (1 + rente/100)  **  x_akse + periodegebyr * maaneder
    fig = plt.figure(figsize=(10,5))
    plt.plot(x_akse, y_akse)
    plt.xlabel('Måneder')
    plt.ylabel('Lån')
    plt.ticklabel_format(useOffset=False, style='plain')
    
    plt.show()

    print("Din endelige sum blir \n (%.2f) " % laan)
```

```python
def annunitets():
    lanemengde = input('Hvor mye vil du låne? \n')
    lanemengde = float(lanemengde.replace(' ', ''))

    terminer = float(input('Hvor mange år vil du ha lånet? \n')) * 12

    rente = float(input('Hva er renten på lånet ditt? \n').replace(',', '.').replace('%', '')) / 100 / 12

    terminbelop = lanemengde * (rente * (1 + rente) ** terminer) / ((1 + rente) ** terminer - 1)


    tabell = []

    termin = 1


    while lanemengde > 0:
        m_rente = rente * lanemengde
        avdrag = terminbelop - m_rente
        lanemengde = lanemengde - avdrag
        tabell.append([termin, terminbelop, m_rente, avdrag, lanemengde])
        termin += 1

    inn = 1
 


    df = pd.DataFrame(tabell, columns=['Termin', 'Terminbeløp', 'Renter', 'Avdrag', 'Restgjeld']).astype(int)
    while (inn != 0 ):
        print(df.truncate(before=inn - 1, after=inn + 3).to_string(index=False))
        inn = int(input('\nHvilken termin vil du se? (0 for å avslutte)\n: '))

       
```

```python
def egenandel():
    kapital = input('Hvor mye koster huset du ønsker å kjøpe?\n')
    kapital = float(kapital.replace(' ', ''))
                            
    egenkapital = kapital * 0.15



    plt.pie([kapital-egenkapital, egenkapital], labels=["kapital", "egenandel"],  explode=(0.05, 0.05))
    plt.title('Husets verdi')
    plt.show()

    print("Egenkapitalen som trengs er\n (%.2f) " % egenkapital) 
    
```

```python

def investeringsprosjekt():
    investeringsbeløp = input('Hvor stort engangs investeringsbeløp?\n')
    investeringsbeløp = float(investeringsbeløp)
    
    rente = input('Hva er forventet avkastning i rente?\n')
    rente = float(rente)
    
    perioder = input('Hvor mange år skal pengene investert?\n')
    perioder = float(perioder) 
    
    fv = investeringsbeløp * (1 + rente/100)**perioder
    
    
    x_akse = np.linspace(0, perioder, int(fv))
    y_akse = investeringsbeløp * (1 + rente/100) ** x_akse
    fig = plt.figure(figsize=(10,5))
    plt.plot(x_akse, y_akse)
    plt.ylabel("Beløp")
    plt.xlabel("År")
    plt.ticklabel_format(useOffset=False, style='plain')
    
    plt.show()
    print("Ditt beløp vil være \n (%.2f) " % fv)
```

```python
def serie():
    lanemengde = input('Hvor mye vil du låne? \n')
    lanemengde = float(lanemengde.replace(' ', ''))

    terminer = int(input('Hvor mange år vil du ha lånet? \n')) * 12

    rente = float(input('Hva er renten på lånet ditt? \n').replace(',', '.').replace('%', '')) / 100 / 12




    avdrag = lanemengde/terminer


    tabell = []

    termin = 1

#viktig for hver iturasjon, så legger vi til inn en rad. legger liste inn med alle kolonnene for den raden.

 
    for termin in range(1, terminer):
        m_rente = rente * lanemengde
        terminbelop = m_rente+avdrag 
        lanemengde = lanemengde - avdrag
        tabell.append([termin, terminbelop, m_rente, avdrag, lanemengde])
       

    inn = 1
 


    df = pd.DataFrame(tabell, columns=['Termin', 'Terminbeløp', 'Renter', 'Avdrag', 'Restgjeld']).astype(int)
    while (inn != 0 ):
        #Du har hele tabellen også tar den fra posisjon eks 0-4 (5 første pos) to_string er til bruk for å ikke vise index- da får du string, som du printer
        #truncate gjør at vi for bare 5 isteden for hele tabellen. 
        print(df.truncate(before=inn - 1, after=inn + 3).to_string(index=False))
        inn = int(input('\nHvilken termin vil du se? (0 for å avslutte)\n: '))
```

```python
while True: 
    inn =input(f'\n1) Annunitetslån\n2) Serielån\n3) Kredittgjeld\n4) Egenandel på huslån\n5) Investeringsprosjekt\n\n0)Avslutt')
    if(inn=='1'):
        annunitets()
    elif(inn == '2'):
        serie()
    elif(inn == '3'):
        kredit()
    elif(inn=='4'):
        egenandel()
    elif(inn == '5'):
        investeringsprosjekt()
    elif(inn == '0'):
        break 
    else:
        print('Ugyldig')
```
